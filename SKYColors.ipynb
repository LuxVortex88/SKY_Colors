```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample dataset
data = {
    'Season': ['Spring', 'Spring', 'Spring', 'Summer', 'Fall', 'Winter', 'Winter'],
    'Time of Day': ['Morning', 'Noon', 'Sunset', 'Noon', 'Sunset', 'Morning', 'Noon'],
    'Sun Elevation (deg)': [15, 70, 10, 85, 5, 10, 30],
    'Sky Color': ['Pale Blue', 'Sky Blue', 'Orange-Red', 'Deep Blue', 'Gold-Red', 'Pale Pink', 'Pale Blue-Grey'],
    'R': [135, 120, 255, 70, 255, 220, 180],
    'G': [206, 180, 120, 130, 100, 190, 200],
    'B': [235, 255, 50, 180, 60, 230, 220]
}

df = pd.DataFrame(data)

# Plot sky colors
plt.figure(figsize=(10, 5))
for i, row in df.iterrows():
    plt.scatter(row['Sun Elevation (deg)'], i,
                color=(row['R']/255, row['G']/255, row['B']/255),
                s=300, label=row['Sky Color'])

plt.yticks(range(len(df)), [f"{s} {t}" for s, t in zip(df['Season'], df['Time of Day'])])
plt.xlabel("Sun Elevation (degrees)")
plt.title("Sky Color by Season and Sun Elevation")
plt.grid(True)
plt.show()
```
